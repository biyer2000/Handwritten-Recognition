```python
import numpy as np
import pandas as pd

train = pd.read_csv('mst/mnist_train.csv')
train
test = pd.read_csv('mst/mnist_test.csv')
test
```

```python
#train = train[train.columns != 'label']
y_train = train['label'].to_numpy()
y_test = test['label'].to_numpy()
```

```python
x_train = train.drop(['label'], axis = 1).to_numpy()
x_test = test.drop(['label'], axis = 1).to_numpy()
```

```python
def one_hot_encode(y_arr):
    new_y = []
    for i in range(len(y_arr)):
        y = np.zeros(10)
        y[y_arr[i]] = 1
        new_y.append(y)
    return np.asarray(new_y)


y_test = one_hot_encode(y_test)
y_train = one_hot_encode(y_train)

```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
class neural_network:
    
    def __init__(self, epochs, lr):
        self.epochs = epochs
        self.lr = lr
        self.w1 = np.random.randn(10, 784) * 0.01
        self.w2 = np.random.randn(10, 10) *0.01
        self.b1 = np.random.randn(10 , 1)
        self.b2 = np.random.randn(10 , 1)
    
    def relu(self, z):
        return np.maximum(0, z)
    
    def softmax(self, z):
        return np.exp(z)/np.sum(np.exp(z))
    def sigmoid(self, z):
        return 1/(1 + np.exp(-z))
    def d_relu(self, z):
        return np.array(z > 0, dtype = np.float32)
    
    def forward_and_backward(self, x, y):
        x = x.reshape(-1, 1)
        y = y.reshape(-1, 1)
        m = y.shape
        z1 = np.dot(self.w1, x) 
        #print(z1)
        a1 = self.relu(z1)
        #print(a1)
        z2 = np.dot(self.w2, a1)
        #print(z2)
        a2 = self.softmax(z2)
        #print(a2)
        dz2 = a2 - y
        #print(y)
        dw2 = np.dot(dz2, a1.T)
        dz1 = np.dot(self.w2.T, dz2) * self.d_relu(a1)
        dw1 = np.dot(dz1, x.T)
        
        self.w1 = self.w1 - self.lr * dw1
        self.w2 = self.w2 - self.lr * dw2
    
    def predict(self, x):
        x = x.reshape(-1, 1)
        z1 = np.dot(self.w1, x)  
        #print(z1)
        a1 = self.relu(z1)
        #print(a1)
        z2 = np.dot(self.w2, a1) 
        #print(z2)
        a2 = self.softmax(z2)
        return a2
    
    def accuracy(self, x_test, y_test):
        count = 0
        for i in range(len(x_test)):
            
            pred = self.predict(x_test[i])
            if np.argmax(pred) == np.argmax(y_test[i].reshape(-1, 1)):
                count += 1
        return count/len(x_test)
    
    def train_and_compute(self, x_train, y_train, x_test, y_test):
        for i in range(self.epochs):
            for j in range(len(x_train)):
                self.forward_and_backward(x_train[j], y_train[j])
            print("accuracy:", self.accuracy(x_test, y_test))
    
    
        
    
     
        
        
```
